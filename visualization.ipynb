```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update({'font.size': 20})
plt.rcParams.update({'lines.linewidth': 2})
```

```python
def genTotSucRate(sucRate):
    totSucRate = np.zeros((len(sucRate),))
    totSucRate[0] = sucRate[0]
    for i in range(1,len(sucRate)):
        totSucRate[i] = totSucRate[i-1] + (1-totSucRate[i-1])*sucRate[i]
    return totSucRate
```

```python
plt.rcParams["figure.figsize"] = (11,5)
raw = np.loadtxt("data/230403/230403_1.csv",delimiter=',',skiprows=1)
totSucRate = genTotSucRate(raw[:,1])
plt.subplot(1,2,1)
plt.plot(raw[:,0],raw[:,1])
plt.subplot(1,2,2)
plt.plot(raw[:,0],totSucRate)
```
